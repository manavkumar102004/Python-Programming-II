# **Assignment - 1 (OOPs Encapsulation )

```python
# Question - 1
class Laptop:
    def __init__(self, price):
        self.__price = price

    def set_price(self, new_price):
        if new_price > 0:
            self.__price = new_price
            print("Price updated successfully.")
        else:
            print("Invalid price.")

    def get_price(self):
        return self.__price

laptop1 = Laptop(60000)
print("Original Price:", laptop1.get_price())
laptop1.set_price(65000)
print("Updated Price:", laptop1.get_price())
```

```python
# Question - 2
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def get_balance(self):
        return self.__balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print("Deposited successfully")
        else:
            print("Invalid amount")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print("Withdrawn successfully")
        else:
            print("Not enough balance")

acc = BankAccount(10000)
acc.deposit(1500)
acc.withdraw(2000)
print("Balance:", acc.get_balance())
```

```python
# Question - 3
class Movie:
    def __init__(self):
        self.__rating = 0

    def set_rating(self, rating):
        if 1 <= rating <= 10:
            self.__rating = rating
            print("Rating set")
        else:
            print("Invalid rating")

    def get_rating(self):
        return self.__rating

movie = Movie()
movie.set_rating(9)
print("Movie Rating:", movie.get_rating())
```

```python
# Question - 4
class Book:
    def __init__(self, title, author):
        self.__title = title
        self.__author = author

    def get_title(self):
        return self.__title

    def get_author(self):
        return self.__author

book1 = Book("Atomic Habits", "James Clear")
print("Title:", book1.get_title())
print("Author:", book1.get_author())
```

```python
# Question - 5
class Car:
    def __init__(self):
        self.__speed = 0

    def increase_speed(self, value):
        if value > 0:
            self.__speed += value
            print("Speed:", self.__speed)

    def decrease_speed(self, value):
        if value > 0:
            self.__speed -= value
            if self.__speed < 0:
                self.__speed = 0
            print("Speed:", self.__speed)

    def get_speed(self):
        return self.__speed

car = Car()
car.increase_speed(120)
car.decrease_speed(50)
print("Final Speed:", car.get_speed())
```
