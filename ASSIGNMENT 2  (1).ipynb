```
# Question 1

class Bike:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed

class ElectricBike(Bike):
    def __init__(self, brand, speed, battery_capacity):
        super().__init__(brand, speed)
        self.battery_capacity = battery_capacity

b1 = ElectricBike("Hero", 80, "5 kWh")
print(b1.brand, b1.speed, b1.battery_capacity)
```

```
# Question 2

class Device:
    def __init__(self, brand, ram):
        self.brand = brand
        self.ram = ram

class Tablet(Device):
    def __init__(self, brand, ram, screen_size):
        super().__init__(brand, ram)
        self.screen_size = screen_size

t1 = Tablet("Samsung", "6GB", "10 inch")
print(t1.brand, t1.ram, t1.screen_size)
```

```
# Question 3

class Gadget:
    def __init__(self, name):
        self.name = name

class Smartwatch(Gadget):
    def __init__(self, name, os):
        super().__init__(name)
        self.os = os

class Headphones(Gadget):
    def __init__(self, name, type_):
        super().__init__(name)
        self.type_ = type_

g1 = Smartwatch("Apple Watch", "watchOS")
g2 = Headphones("Boat", "Wireless")

print(g1.name, g1.os)
print(g2.name, g2.type_)
```

```
# Question 4

class Artist:
    def __init__(self, name):
        self.name = name

class Painter(Artist):
    def __init__(self, name, style):
        super().__init__(name)
        self.style = style

p1 = Painter("Ravi Verma", "Oil Painting")
print(p1.name, p1.style)
```

```
# Question 5

class Book:
    def __init__(self, title):
        self.title = title

class ScienceBook(Book):
    def __init__(self, title, field):
        super().__init__(title)
        self.field = field

b1 = ScienceBook("Physics Basics", "Physics")
print(b1.title, b1.field)
```

```
# Bonus Question

class Student:
    def __init__(self, name, m1, m2, m3):
        self.name = name
        self.m1 = m1
        self.m2 = m2
        self.m3 = m3

    def average(self):
        avg = (self.m1 + self.m2 + self.m3) / 3
        print("Average Marks:", avg)

s1 = Student("Anup", 75, 85, 90)
s1.average()
```
