## 🔹 Iterators

### Example 1: Iterator using list

```python
nums = [10, 20, 30, 40, 50]
it = iter(nums)

print(next(it))
print(next(it))
```

### Example 2: Iterator with string

```python
text = "World"
text_iter = iter(text)

print(next(text_iter))
print(next(text_iter))
```

### Example 3: Using for loop

```python
data = [5, 10, 15, 20]

it = iter(data)
for value in it:
    print(value)
```

### Example 4: Iterator with range

```python
r = iter(range(3, 8))

print(next(r))
print(next(r))
```

### Example 5: While loop with iterator

```python
colors = ("red", "green", "blue")
color_iter = iter(colors)

while True:
    try:
        print(next(color_iter))
    except StopIteration:
        break
```

## 🔹 Decorators

### Example 1: Basic decorator

```python
def simple_decorator(func):
    def wrapper():
        print("Starting...")
        func()
    return wrapper

@simple_decorator
def greet():
    print("Good Morning!")

greet()
```

### Example 2: Decorator with lines

```python
def border(func):
    def wrapper():
        print("********")
        func()
        print("********")
    return wrapper

@border
def message():
    print("Welcome!")

message()
```

### Example 3: Before and After decorator

```python
def check(func):
    def wrapper():
        print("Before execution")
        func()
        print("After execution")
    return wrapper

@check
def say_hi():
    print("Hi!")

say_hi()
```

### Example 4: Timer decorator

```python
import time

def timer(func):
    def wrapper():
        start = time.time()
        func()
        end = time.time()
        print("Execution time:", end - start)
    return wrapper

@timer
def task():
    time.sleep(2)
    print("Task completed")

task()
```

### Example 5: Access control decorator

```python
def check_user(func):
    def wrapper(user):
        if user != "admin":
            print("Not allowed")
            return
        func(user)
    return wrapper

@check_user
def dashboard(user):
    print(user, "entered dashboard")

dashboard("guest")
dashboard("admin")
```

### Example 6: Factorial with caching

```python
def cache_decorator(func):
    store = {}
    def wrapper(n):
        if n in store:
            return store[n]
        result = func(n)
        store[n] = result
        return result
    return wrapper

@cache_decorator
def fact(n):
    if n == 0:
        return 1
    return n * fact(n-1)

print(fact(6))
```

## 🔹 Generators

### Example 1: Simple generator

```python
def gen_func():
    yield "AI"
    yield "ML"
    yield "DS"

g = gen_func()
print(next(g))
```

### Example 2: Looping generator

```python
def numbers():
    yield 4
    yield 5
    yield 6

g = numbers()
for i in g:
    print(i)
```

### Example 3: Squares using generator expression

```python
sq = (x**2 for x in range(20))

print(next(sq))
print(next(sq))
print(next(sq))
```

### Example 4: Fibonacci generator

```python
def fib(n):
    a, b = 1, 1
    for i in range(n):
        yield a
        a, b = b, a + b

f = fib(10)
print(list(f))
```

### Example 5: Generator with send()

```python
def counter(n):
    while n > 0:
        val = (yield n)
        if val:
            n = val
        else:
            n -= 1

c = counter(4)
print(next(c))
print(c.send(2))
print(next(c))
```
